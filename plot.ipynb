```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df = pd.read_parquet('uk_polling_data.parquet')
df = df.dropna(subset=['date'])
```

```python
plt.figure(figsize=(12, 6))
parties = ['conservative', 'labour', 'reform', 'liberal_democrats', 'green', 'plaid_cymru', 'snp', 'others']
party_colors = {
    'conservative': 'blue',
    'labour': 'red',
    'reform': 'cyan',
    'liberal_democrats': 'orange',
    'green': 'green',
    'plaid_cymru': 'darkgreen',
    'snp': 'yellow',
    'others': 'gray'
}

for party in parties:
    df.plot.scatter(x='date', y=party, ax=plt.gca(), label=party, color=party_colors.get(party, 'black'), edgecolor='gray', alpha=0.7, s=15)
```
